```python
# data manipulation
import pandas as pd

# visualization
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style('whitegrid')

# importing warnings library to ignore any kind of irrelevant warning
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
# reading file (the file can be found within this repository)
dmds = pd.read_csv('5 year comparison DMDS.csv')

# looking at the first 5 rows of the dataframe
dmds.head()
```

```python
# transforming the column 'week' into datetime format
dmds['Week'] = pd.to_datetime(dmds['Week'])

# setting 'week' as index
dmds.set_index('Week', inplace=True)

# looking again at the first 5 rows of the transformed dataframe
dmds.head()
```

## Google Trends comparison: Digital Marketing & Data Science

By gathering data from Google Trends of Digital Marketing & Data Science, and analyzing it with Python we can build a quick visualization of the evolution of interest over time* of both terms from 2014 until 2018.

*For further information about "Interest over time" please visit Google Trends.

```python
def google_trends_plot():    
    
    # resetting index
    dmds.reset_index(inplace=True)
    
    #canvas
    dmds.plot(x='Week', figsize=(12, 8), grid=True)
    
    #title
    plt.title('Google Trends comparison: Digital Marketing & Data Science', fontsize=14)  
    
    #x & y labels
    plt.ylabel('Interest over time', fontsize=14, )
    plt.xlabel('Data per week for the past 5 years', fontsize=14)
    
    #x limits
    plt.xlim(dmds['Week'].iloc[0], dmds['Week'].iloc[-1])
    
    #tight layout
    plt.tight_layout()
```

The amount of insights that a digital marketing tool such as Google Trends can provide for things like SEO (Search Engine Optimization) and Ad Campaigns is countless, nevertheless, such tools are limited as they do not count with the statistical capabilities that Data Science provides. This is where Digital Marketing collides with Data Science. Talking about Keyword Search Queries automatically links to Data Analytics.

```python
google_trends_plot()
```

## Linear Regression plot

Take this Linear Regression plot as an example, which is a visualization that Google Trends does not offer but the data analysis capabilities of Python does.

```python
def linear_regression_plot():
    sns.lmplot(data=dmds, x='Data Science: (Worldwide)', y='Digital Marketing: (Worldwide)', aspect=10/8)

    #title
    plt.title('Linear Regression plot', fontsize=14)

    #labels
    plt.xlabel('Digital Marketing', fontsize=14)
    plt.ylabel('Data Science', fontsize=14)

    #tight layout
    plt.tight_layout()
```

```python
linear_regression_plot()
```
